## DATA CORRELATION AMONG COLUMNS

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
plt.imshow(df.corr(), cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(df.corr())), df.corr().columns, rotation=90)
plt.yticks(range(len(df.corr())), df.corr().columns)
plt.show()
```

```python
df.plot(kind='bar', x='pH', y='alcohol')

plt.title('Chart of pH against alcohol')
plt.xlabel('alcohol')
plt.ylabel('pH')
plt.xticks(rotation=45)

plt.show()
```
